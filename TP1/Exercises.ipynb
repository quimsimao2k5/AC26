### 5. Factorial
Ask the user for a number `n` and compute `n!` (factorial).

Provide two solutions:
A) One using a for loop.
B) One using a while loop. (Do not use recursion)

- $1! = 1$
- $2! = 2 \times 1 = 2$
- $3! = 3 \times 2 \times 1 = 6$
- $4! = 4 \times 3 \times 2 \times 1 = 24$
- $5! = 5 \times 4 \times 3 \times 2 \times 1 = 120$

```python
def fatorialFor(n):
  fact=1
  for i in range(2,n+1):
    fact*=i
  return fact

def fatorialWhile(n):
  fact=1
  i=2
  while(i<=n):
    fact*=i
    i+=1
  return fact

def fatorialRec(n):
  if n in [0,1]:
    return 1
  elif n>1:
    return n*fatorialRec(n-1)

n = int(input("Fatorial de "))
print("For: ",fatorialFor(n))
print("While: ",fatorialWhile(n))
print("Rec: ",fatorialRec(n))
```

### 1. Palindrome
Write `is_palindrome(word)` returning `True` if `word` reads the same backwards.  

```python
def is_palindrome(s):
  s = str(s).strip().lower()
  return s == s[::-1]

print(is_palindrome(input("Manda uma palavra ")))
```

### 2. Normalize
Write a function `normalize(xs)` that takes a list of numbers and returns a new list scaled to sum to 1.
- If the sum is 0, return a list of zeros of the same length.

```python
def normalize(xs):
  s=sum(xs)
  if s==0:
    return [0 for _ in xs]
  else:
    return [x/s for x in xs]
print(normalize([1,2,3]))
```

### 3. Frequency table
Implement `freq(s)` returning a dict `{char: count}` for a string.

```python
def freq(s):
    d={}
    for l in "".join(s.lower().split(" ")):
        d[l] = d.get(l,0)+1
    return d

print(freq("Carla mendes algebra universal"))
```
